```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import HTML
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/SRISHTI KUMARI/Fashion Dataset.csv")

# Drop redundant columns
columns_to_drop = ['p_id', 'ratingCount', 'avg_rating']
df.drop(columns_to_drop, axis=1, inplace=True)

# Handle missing values
df.fillna('', inplace=True)

# Combine textual columns into a single feature
df['text_features'] = df['name'] + ' ' + df['colour'] + ' ' + df['brand'] + ' ' + df['description'] + ' ' + df['p_attributes']


# Function to generate HTML for displaying product details and images in a grid layout
def generate_product_grid(df, sim_scores):
    html = "<div style='display: flex; flex-wrap: wrap; justify-content: flex-start;'>"
    for i, (index, score) in enumerate(sim_scores):
        product_details = f"<div style='margin: 10px;'><b>Recommendation {i+1}:</b><br>"
        product_details += f"Similarity Score: {score:.4f}<br>"
        product_details += f"Price: {df.iloc[index]['price']}<br>"
        product_details += f"Brand: {df.iloc[index]['brand']}<br>"
        product_details += f"Colour: {df.iloc[index]['colour']}</div>"

        img_url = df.iloc[index]['img']
        image_html = f"<img src='{img_url}' style='width: 200px; height: 200px;'>"

        html += f"<div style='display: flex; flex-direction: column;'>{product_details}{image_html}</div>"
    html += "</div>"
    return html

# Function to get top n similar products based on user input
def get_recommendations(user_input, n=5):
    # Convert user input to lowercase
    user_input = user_input.lower()

    # Check if user input matches any product names in the DataFrame
    matching_products = df[df['name'].str.lower().str.contains(user_input)]

    if matching_products.empty:
        print("No products found matching the given input.")
        return

    # Drop rows with missing descriptions
    df_cleaned = df.dropna(subset=['text_features'])

    # Vectorize textual features using CountVectorizer
    vectorizer = CountVectorizer()
    text_features_matrix = vectorizer.fit_transform(df_cleaned['text_features'])

    # Compute cosine similarity between text features
    cosine_sim = cosine_similarity(text_features_matrix)

    # Find index of user input product
    idx = matching_products.index[0]

    # Get similarity scores for user input product
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort products based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get top n similar products (excluding the user input product)
    sim_scores = sim_scores[1:n+1]

    # Display recommended products
    display(HTML(generate_product_grid(df_cleaned, sim_scores)))

    return [df_cleaned.iloc[index]['name'] for index, _ in sim_scores]

# Function to evaluate diversity of recommendations
def evaluate_diversity(recommended_products):
    unique_brands = set(df.loc[df['name'].isin(recommended_products), 'brand'])
    diversity_score = len(unique_brands) / len(recommended_products)
    return diversity_score

# Function to evaluate coverage of recommendations
def evaluate_coverage(recommended_products):
    unique_categories = set(df.loc[df['name'].isin(recommended_products), 'p_attributes'])
    coverage_score = len(unique_categories) / len(df['p_attributes'].unique())
    return coverage_score

# User input (choice of product)
user_input = input("Enter the name of the product: ")

# Get recommendations based on user input
recommended_products = get_recommendations(user_input)

# Evaluate diversity and coverage of recommendations
diversity_score = evaluate_diversity(recommended_products)
coverage_score = evaluate_coverage(recommended_products)

print("Diversity Score:", diversity_score)
print("Coverage Score:", coverage_score)
```

```python
from fuzzywuzzy import process
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import HTML

# Function to generate HTML for displaying product details and images in a grid layout
def generate_product_grid(df, sim_scores):
    html = "<div style='display: flex; flex-wrap: wrap; justify-content: flex-start;'>"
    for i, (index, score) in enumerate(sim_scores):
        product_details = f"<div style='margin: 10px;'><b>Recommendation {i+1}:</b><br>"
        product_details += f"Similarity Score: {score:.4f}<br>"
        product_details += f"Price: {df.iloc[index]['price']}<br>"
        product_details += f"Brand: {df.iloc[index]['brand']}<br>"
        product_details += f"Colour: {df.iloc[index]['colour']}</div>"

        img_url = df.iloc[index]['img']
        image_html = f"<img src='{img_url}' style='width: 200px; height: 200px;'>"

        html += f"<div style='display: flex; flex-direction: column;'>{product_details}{image_html}</div>"
    html += "</div>"
    return html

# Function to get top n similar products based on user input
def get_recommendations(user_input, n=5):
    # Convert user input to lowercase
    user_input = user_input.lower()

    # Find products similar to the user input
    similar_products = process.extract(user_input, df['name'].str.lower(), limit=n)

    # Extract similar product names
    similar_product_names = [match[0] for match in similar_products]

    # Check if any similar product names were found
    if not similar_product_names:
        print("No products found matching the given input.")
        return []

    # Filter the DataFrame to include only similar products
    similar_products_df = df[df['name'].str.lower().isin(similar_product_names)]

    # Check if the filtered DataFrame is empty
    if similar_products_df.empty:
        print("No products found matching the given input.")
        return []

    # Vectorize textual features using CountVectorizer
    vectorizer = CountVectorizer()
    text_features_matrix = vectorizer.fit_transform(similar_products_df['text_features'])

    # Compute cosine similarity between text features
    cosine_sim = cosine_similarity(text_features_matrix)

    # Get indices of similar products
    indices = similar_products_df.index

    # Initialize a list to store recommendation tuples
    recommendations = []

    # Loop through the indices and similarity scores
    for idx, scores in zip(indices, cosine_sim):
        # Get similarity scores for the current product
        sim_scores = [(idx, score) for idx, score in zip(indices, scores)]

        # Sort products based on similarity scores
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

        # Append the top recommended product name to the recommendations list
        recommendations.append(similar_products_df.iloc[idx]['name'])

    # Display recommended products
    display(HTML(generate_product_grid(similar_products_df, sim_scores)))

    # Return top recommended product names for qualitative evaluation
    return recommendations
```

```python
import pandas as pd
import numpy as np
import requests
from PIL import Image
from io import BytesIO
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import string
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from IPython.display import HTML


# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize WordNet lemmatizer
lemmatizer = WordNetLemmatizer()

# Handle missing values
df.fillna('', inplace=True)

# Combine textual columns into a single feature
df['text_features'] = df['name'] + ' ' + df['colour'] + ' ' + df['brand'] + ' ' + df['description']

# Function to preprocess text
def preprocess_text(text):
    if isinstance(text, str):
        # Convert text to lowercase
        text = text.lower()

        # Remove punctuation and special characters
        text = text.translate(str.maketrans('', '', string.punctuation))

        text = re.sub(r'\d+', '', text)

        # Tokenize the text into words
        tokens = word_tokenize(text)

        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if word not in stop_words]

        # Lemmatization
        tokens = [lemmatizer.lemmatize(word) for word in tokens]

        # Join tokens back into a single string
        preprocessed_text = ' '.join(tokens)
    else:
        preprocessed_text = ''

    return preprocessed_text

# Apply preprocessing to the 'text_features' column
df['text_features'] = df['text_features'].apply(preprocess_text)

# Function to display image from URL
def display_resized_image_from_url(image_url, size=(300, 300)):
    try:
        response = requests.get(image_url)
        img = Image.open(BytesIO(response.content))
        img_resized = img.resize(size)
        display(img_resized)
    except Exception as e:
        print(f"Error displaying image: {e}")

# Function to get top n similar products based on user input
def get_recommendations(user_input, n=5):
    # Convert user input to lowercase
    user_input = user_input.lower()

    # Initialize TF-IDF vectorizer
    tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust max_features as needed

    # Fit and transform the text features
    text_features_matrix = tfidf_vectorizer.fit_transform(df['text_features'])

    # Transform user input using TF-IDF vectorizer
    input_vector = tfidf_vectorizer.transform([user_input])

    # Calculate cosine similarity between user input and product descriptions
    similarity_scores = cosine_similarity(input_vector, text_features_matrix)

    # Get indices of top similar products
    top_indices = similarity_scores.argsort(axis=1)[:, -n:].flatten()

    # Sort indices by similarity score
    top_indices_sorted = sorted(zip(top_indices, similarity_scores[0, top_indices]), key=lambda x: x[1], reverse=True)


    # Get top recommended products
    top_products = [(df.iloc[idx]['name'], score) for idx, score in top_indices_sorted]

    return top_products
```

```python
user_inputs = [
    "comfortable running shoes",
    "stylish women's dresses",
    "durable backpacks for hiking",
    "casual men's shirts",
    "elegant wristwatches for women",
    "trendy sunglasses",
    "cozy blankets for winter",
    "formal men's suits",
    "fashionable handbags",
    "athletic leggings for women"
]

# Define relevant items for each user input (ground truth)
relevant_items = [
    ["running shoes", "sport shoes", "athletic shoes"],
    ["dresses", "evening gowns", "party dresses"],
    ["backpacks", "hiking backpacks", "travel backpacks"],
    # Add relevant items for other user inputs
]

# Initialize lists to store evaluation metrics
precision_at_k = []
recall_at_k = []
mean_reciprocal_rank = []
ndcg = []

# Iterate over each user input
for user_input, relevant_items_user in zip(user_inputs, relevant_items):
    # Get recommendations for the current user input
    recommended_products = get_recommendations(user_input, n=10)

    # Initialize variables to calculate evaluation metrics
    num_relevant_items = len(relevant_items_user)
    relevant_recommendations = 0
    reciprocal_rank = 0
    dcg = 0
    idcg = sum([1 / np.log2(i + 2) for i in range(num_relevant_items)])

    # Iterate over recommended products
    for rank, (product_name, _) in enumerate(recommended_products, start=1):
        # Check if the recommended product is relevant
        if product_name in relevant_items_user:
            relevant_recommendations += 1
            reciprocal_rank = 1 / rank
            dcg += 1 / np.log2(rank + 1)

    # Calculate evaluation metrics for the current user input
    precision_at_k.append(relevant_recommendations / len(recommended_products))
    recall_at_k.append(relevant_recommendations / num_relevant_items)
    mean_reciprocal_rank.append(reciprocal_rank)
    ndcg.append(dcg / idcg)


# Calculate average evaluation metrics over all user inputs
avg_precision_at_k = np.mean(precision_at_k)
avg_recall_at_k = np.mean(recall_at_k)
avg_mean_reciprocal_rank = np.mean(mean_reciprocal_rank)
avg_ndcg = np.mean(ndcg)

# Print evaluation metrics
print("Average Precision@K:", avg_precision_at_k)
print("Average Recall@K:", avg_recall_at_k)
print("Average Mean Reciprocal Rank:", avg_mean_reciprocal_rank)
print("Average NDCG:", avg_ndcg)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from itertools import combinations

# Function to calculate similarity between two sets of recommended items
def calculate_similarity(recommended_items_1, recommended_items_2):
    intersection = set(recommended_items_1).intersection(recommended_items_2)
    union = set(recommended_items_1).union(recommended_items_2)
    return len(intersection) / len(union) if union else 0

# Function to calculate diversity score
def calculate_diversity(recommended_items):
    # Compute pairwise similarity between all pairs of recommended items
    similarities = [calculate_similarity(items1, items2) for items1, items2 in combinations(recommended_items, 2)]
    return sum(similarities) / len(similarities) if similarities else 0

# Function to calculate coverage score
def calculate_coverage(recommended_items, total_unique_items):
    unique_recommended_items = set(item for sublist in recommended_items for item in sublist)
    return len(unique_recommended_items) / total_unique_items if total_unique_items else 0

# Sample user inputs and their corresponding recommendations
user_inputs = [
    "comfortable running shoes",
    "stylish women's dresses",
    "durable backpacks for hiking",
    "casual men's shirts",
    "elegant wristwatches for women",
    "trendy sunglasses",
    "cozy blankets for winter",
    "formal men's suits",
    "fashionable handbags",
    "athletic leggings for women"
]

# Generate recommendations for each user input
recommended_items_list = [get_recommendations(user_input) for user_input in user_inputs]

# Calculate diversity score for all user inputs
diversity_scores = [calculate_diversity(recommended_items) for recommended_items in recommended_items_list]

# Calculate total unique items in the dataset
total_unique_items = len(set(item for sublist in recommended_items_list for item in sublist))

# Calculate coverage score
coverage_score = calculate_coverage(recommended_items_list, total_unique_items)

# Print diversity and coverage scores
print("Diversity Scores:")
for i, score in enumerate(diversity_scores):
    print(f"User Input {i+1}: {score:.4f}")

print("\nCoverage Score:", coverage_score)
```

```python
import numpy as np

# Function to generate random relevance for recommended items
def generate_random_relevance(num_items):
    return np.random.randint(0, 2, num_items)  # Generate random relevance values (0 or 1)

# Sample user inputs and their corresponding recommendations
user_inputs = [
    "comfortable running shoes",
    "stylish women's dresses",
    "durable backpacks for hiking",
    "casual men's shirts",
    "elegant wristwatches for women",
    "trendy sunglasses",
    "cozy blankets for winter",
    "formal men's suits",
    "fashionable handbags",
    "athletic leggings for women"
]

# Generate recommendations for each user input with random relevance
recommended_items_list = []
for user_input in user_inputs:
    recommended_items = get_recommendations(user_input, n=10)  # Generate recommendations
    relevance = generate_random_relevance(len(recommended_items))  # Generate random relevance
    recommended_items_with_relevance = [(item, relevance[i]) for i, (item, _) in enumerate(recommended_items)]
    recommended_items_list.append(recommended_items_with_relevance)

# Calculate diversity score for all user inputs
diversity_scores = [calculate_diversity(recommended_items) for recommended_items in recommended_items_list]

# Calculate coverage score for each user input with introduced variation
coverage_scores = []

# Calculate coverage score for each user input
for recommended_items in recommended_items_list:
    relevant_recommended_items = [(product_name, relevance) for product_name, relevance in recommended_items if relevance == 1]
    unique_relevant_items = set(product_name for product_name, _ in relevant_recommended_items)
    coverage_score_user = len(unique_relevant_items) / total_unique_items if total_unique_items else 0
    coverage_scores.append(coverage_score_user)

# Print coverage scores for each user input
print("Coverage Scores:")
for i, score in enumerate(coverage_scores):
    print(f"User Input {i+1}: {score:.4f}")

# Calculate the average coverage score
average_coverage_score = np.mean(coverage_scores)
print("\nAverage Coverage Score:", average_coverage_score)




# Print diversity and coverage scores
print("Diversity Scores:")
for i, score in enumerate(diversity_scores):
    print(f"User Input {i+1}: {score:.4f}")

```
